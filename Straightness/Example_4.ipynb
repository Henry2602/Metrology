```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("grayscale")
```

```python
x_value = np.array([0, 25.4, 50.8, 76.2, 101.6, 127, 152.4, 177.8, 203.2, 228.6, 254, 279.4, 304.8, 330.8, 355.6, 381, 406.4, 431.8, 457.2, 482.6, 508, 533.4])
y_value = np.array([0.41, 0, -0.108, -0.17, -0.112, -0.068, -0.05, -0.15, -0.302, -0.286, -0.22, -0.18, -0.148, -0.078, -0.178, -0.22, -0.272, -0.334, -0.266, -0.126, 0, 0.15])
z_value = np.array([0, 0.124, 0.205, 0.306, 0.352, 0.387, 0.326, 0.248, 0.256, 0.1, -0.0688, -0.262, -0.558, -0.74, -0.899, -0.956, -0.942, -0.928, -0.88, -0.98, -1.04, -0.83])
y_0 = np.zeros((len(z_value),1))
z_0 = np.zeros((len(z_value),1))
```

```python
A = np.column_stack((np.ones(len(x_value)), x_value))
y = np.transpose(y_value)
z = np.transpose(z_value)
```

```python
b = np.dot(np.linalg.inv(np.dot(np.transpose(A), A)),np.dot(np.transpose(A), y))
print(b)
b_z = np.dot(np.linalg.inv(np.dot(np.transpose(A), A)),np.dot(np.transpose(A), z))
print(b_z)
```

```python
ax = plt.figure(figsize=(6, 6)).add_subplot(projection="3d")
ax.plot(x_value, y_0, z_value, "g.")
ax.plot(x_value, y_value, "b.")
ax.plot(x_value, b[0] + b[1]*x_value, color="orange")
ax.plot(x_value, y_0, b_z[0] + b_z[1] * x_value, color="red")
ax.legend(["points in (x, y)", "points in (x, z)", "line in (x, z)", "line in (x, y)"])
ax.set_xlabel("x (cm)")
ax.set_ylabel('y (μm)')
ax.set_zlabel('z (μm)')
plt.show()
```

```python
x_y_plot = plt.figure().add_subplot()
x_y_plot.plot(x_value, y_value, "g.")
x_y_plot.plot(x_value, b[0] + b[1]*x_value, "r")
```

```python
x_z_plot = plt.figure().add_subplot()
x_z_plot.plot(x_value, z_value, "b.")
x_z_plot.plot(x_value, b_z[0] + b_z[1]*x_value, "orange")
```
