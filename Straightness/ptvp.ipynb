```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
t_start = 0
t_stop = 240
increment = 0.01
x_inital = [0, 0]
t = np.arange(t_start, t_stop, increment)
```

```python
def mydiff(x, t): 
    c = 4 #Damping Constant
    k =1 #Stiffness of the spring
    F = 5
    m = 10 #mass
    dx1dt = x[1]
    dx2dt = (F-c*x[1]-k*x[0])/m
    dxdt = [dx1dt, dx2dt]
    return dxdt
```

```python
x = odeint(mydiff, x_inital, t)
x1 = x[:, 0]
x2 = x[:,1]
print(np.shape(x2))
```

```python
plt.plot(t, x1)
plt.plot(t, x2)
```
